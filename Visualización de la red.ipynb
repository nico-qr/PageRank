```python
import pickle
with open("datos_save.pk1", "rb") as f:
    data = pickle.load(f)
```

```python
print(len(data))
```

```python
nempt = [(key, value) for key, value in data.items() if value is not None]
```

```python
print(len(nempt))
```

```python
pem = [par for par in nempt if len(par[1])>100]
```

```python
print(len(pem))
```

```python
import networkx as nx
```

```python
G = nx.Graph()
```

```python
import matplotlib.pyplot as plt
```

```python

# Your list
#my_list = [(1, (2,)), (2, (1, 3, 4)), (3, (1, 2, 4)), (4, (2, 3))]

my_list = pem

# Create a directed graph
G = nx.DiGraph()

# Add nodes with the first elements of each tuple
G.add_nodes_from([item[0] for item in my_list])

# Add edges connecting each vertex to its adjacent vertices
for item in my_list:
    vertex = item[0]
    adjacent_vertices = item[1]
    G.add_edges_from([(vertex, adj_vertex) for adj_vertex in adjacent_vertices])
    

pos = nx.spring_layout(G)
```

```python
nx.draw_networkx_nodes(G, pos, cmap=plt.get_cmap('jet'), node_size = 500)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos, edgelist=G.edges, edge_color='r', arrows=True)
plt.show()
```
